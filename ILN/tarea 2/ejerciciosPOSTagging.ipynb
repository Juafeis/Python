1.- Analiza el corpus etiquetado Brown, utilizando el conjunto de etiquetas 'universal' (tagset='universal'). En concreto, se desea…

```python
import pprint
import nltk
import matplotlib.pylab as plt
from nltk.tag import UnigramTagger, BigramTagger, TrigramTagger
from nltk.corpus import brown, treebank
nltk.download('brown')

brown.words()
treebank.words()

tagged = []

for category in brown.categories(): 
    tagged.append(len(brown.tagged_words(tagset = 'universal', categories = category)))
plt.plot(brown.categories(),tagged)
plt.legend(loc="lower left", mode = "expand", ncol = len(brown.categories()))
plt.xticks(rotation=90)
```

```python
tagged = []
for tag in brown.tagged_sents():
    tagged.append(len(brown.tagged_sents(tagset = tag)))
    
plt.plot(brown.tagged_sents(),tagged)
```
